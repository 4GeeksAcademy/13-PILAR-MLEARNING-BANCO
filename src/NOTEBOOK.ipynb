# ANALISIS DE VARIABLES CATEGÓRICAS

````python
# ANALISIS DE VARIABLES CATEGORICAS
import matplotlib.pyplot as plt 
import seaborn as sns
import pandas as pd

# Cargar los datos
DATOS = '/workspaces/13-PILAR-MACHINELEARNING/DATOS.csv'
df = pd.read_csv("https://raw.githubusercontent.com/4GeeksAcademy/logistic-regression-project-tutorial/main/bank-marketing-campaign-data.csv", delimiter=';')

# Identificar las columnas categóricas
columnas_categoricas = ['job', 'marital', 'education', 'default', 'housing', 'loan', 'contact', 'month', 'day_of_week', 'poutcome']

# Crear subplots dinámicamente basado en el número de columnas categóricas
fig, axis = plt.subplots(2, 5, figsize =(20,8))  # Ajustar el tamaño según el número de columnas
axis = axis.flatten()  # Aplanar la matriz de ejes para un acceso más fácil

# Crear un histograma múltiple solo para las variables categóricas
for i, col in enumerate(columnas_categoricas):
    if col in df.columns:
        sns.histplot(ax = axis[i], data = df, x = col).set(ylabel = None)
        axis[i].set_title(col)

# Ajustar el layout
plt.tight_layout()
# Mostrar el plot
plt.show()
````

- job (trabajo):Predominan los trabajos en servicios, administración y técnicos.
- marital (estado civil): La mayoría de los individuos están casados, seguidos por solteros y divorciados.
- education (educación): Principalmente, los individuos tienen educación secundaria y universitaria.
- default (incumplimiento): La mayoría de los individuos no tienen créditos en incumplimiento.
- housing (vivienda): Hay una distribución equilibrada entre quienes tienen y no tienen préstamos hipotecarios.
- loan (préstamo personal):La mayoría de los individuos no tienen préstamos personales.
- contact (método de contacto):La mayoría de los contactos se realizaron por teléfono.
- month (mes):La mayoría de los contactos se realizaron en mayo.
- day_of_week (día de la semana):Los contactos están distribuidos de manera uniforme a lo largo de la semana, con picos los lunes y martes.
- poutcome (resultado de la campaña anterior):La mayoría de los individuos no participaron en la campaña anterior.

Conclusión
La mayoría de los individuos están casados, trabajan en servicios o administración, y tienen educación secundaria o universitaria. Los contactos se realizaron principalmente por teléfono y en mayo. La mayoría no participaron en la campaña anterior.

# ANALISIS DE VARIABLES MULTIVARIANTE

VARIABLES NUMERICAS - NUMERICAS

````python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
df = pd.read_csv("https://raw.githubusercontent.com/4GeeksAcademy/logistic-regression-project-tutorial/main/bank-marketing-campaign-data.csv", sep = ";")
# Identificar las columnas numéricas y la variable objetivo
columnas_numericas = ['age', 'duration', 'campaign', 'pdays', 'previous', 'emp.var.rate', 'cons.price.idx', 'cons.conf.idx', 'euribor3m', 'nr.employed']
target = 'y'
# Convertir la variable objetivo a valores numéricos
df[target] = df[target].apply(lambda x: 1 if x == 'yes' else 0)
# Crear subplots dinámicamente basado en el número de variables numéricas
num_plots = len(columnas_numericas)
fig, axis = plt.subplots((num_plots + 3) // 4, 4, figsize=(20, (num_plots + 3) // 4 * 5))
# Aplanar la matriz de ejes para un acceso más fácil
axis = axis.flatten()
# Crear un diagrama de dispersión para cada variable numérica con la variable objetivo `y`
for i, col in enumerate(columnas_numericas):
    sns.regplot(ax=axis[i], data=df, x=col, y=target).set(ylabel=None)
    axis[i].set_title(f'{col} vs {target}')
# Eliminar los ejes vacíos
for j in range(len(columnas_numericas), len(axis)):
    fig.delaxes(axis[j])
# Ajustar el layout
plt.tight_layout()
# Mostrar el plot
plt.show()
````

### Análisis de la Relación entre Variables Numéricas y la Variable Objetivo `y`

1. **age (edad) vs y**: La edad no muestra una relación fuerte con la probabilidad de suscripción.
2. **duration (duración) vs y**:Llamadas más largas tienden a resultar en una mayor tasa de suscripción.
3. **campaign (campaña) vs y**:El número de contactos durante la campaña no está fuertemente relacionado con la suscripción.
4. **pdays (días desde la última campaña) vs y**:Los días desde la última campaña no muestran una relación fuerte con la suscripción.
5. **previous (contactos previos) vs y**:El número de contactos previos no está fuertemente relacionado con la suscripción.
6. **emp.var.rate (tasa de variación del empleo) vs y**:La tasa de variación del empleo muestra una relación moderada con la suscripción.
7. **cons.price.idx (índice de precios al consumidor) vs y**:El índice de precios al consumidor tiene una relación moderada con la suscripción.
8. **cons.conf.idx (índice de confianza del consumidor) vs y**: La confianza del consumidor muestra una relación moderada con la suscripción.
9. **euribor3m (tasa Euribor a 3 meses) vs y**:La tasa Euribor a 3 meses tiene una relación significativa con la suscripción.
10. **nr.employed (número de empleados) vs y**:El número de empleados muestra una relación moderada con la suscripción.

Las variables económicas como la duración de las llamadas, la tasa de variación del empleo, el índice de precios al consumidor, la tasa Euribor a 3 meses y el número de empleados muestran una relación moderada a alta con la probabilidad de suscripción de un depósito a plazo. Estas relaciones sugieren que las condiciones económicas y la duración de las interacciones tienen un impacto notable en la decisión de suscribir. Por otro lado, variables como la edad, el número de contactos durante la campaña y los días desde la última campaña no muestran una correlación fuerte con la probabilidad de suscripción.

````python
import pandas as pd

import json

total_data_con_outliers = total_data.copy()
total_data_sin_outliers = total_data.copy()

def replace_outliers_from_column(column, df):
  column_stats = df[column].describe()
  column_iqr = column_stats["75%"] - column_stats["25%"]
  upper_limit = column_stats["75%"] + 1.5 * column_iqr
  lower_limit = column_stats["25%"] - 1.5 * column_iqr
  if lower_limit < 0: lower_limit = min(df[column])
  # Remove upper outliers
  df[column] = df[column].apply(lambda x: x if (x <= upper_limit) else upper_limit)
  # Remove lower outliers
  df[column] = df[column].apply(lambda x: x if (x >= lower_limit) else lower_limit)
  return df.copy(), [lower_limit, upper_limit]

outliers_dict = {}
for column in ["pdays", "previous", "duration"]:
  total_data_sin_outliers, limits_list = replace_outliers_from_column(column, total_data_sin_outliers)
  outliers_dict[column] = limits_list

# with open("outliers_replacement.json", "w") as f:
#   json.dump(outliers_dict, f)
outliers_dict
````

# OPTIMIZACION DEL MODELO

#### meto hiperparametros : El código utiliza **GridSearchCV** para encontrar los mejores **hiperparámetros** para el modelo de regresión logística. Los hiperparámetros son configuraciones que puedes ajustar para mejorar el rendimiento del modelo.
#### **Definición de los hiperparámetros**
```python
hyperparams = {
    "C": [0.001, 0.01, 0.1, 1, 10, 100, 1000],
    "penalty": ["l1", "l2", "elasticnet", None],
    "solver": ["newton-cg", "lbfgs", "liblinear", "sag", "saga"]
}
```
- **`C`**: Controla la regularización del modelo. Valores más pequeños significan más regularización (menos sobreajuste).
- **`penalty`**: Define el tipo de regularización que se aplicará:
  - `"l1"`: Regularización Lasso.
  - `"l2"`: Regularización Ridge.
  - `"elasticnet"`: Combinación de L1 y L2.
  - `None`: Sin regularización.
- **`solver`**: Es el algoritmo que se usa para ajustar el modelo. Algunos ejemplos:
  - `"newton-cg"` y `"lbfgs"`: Métodos rápidos para problemas grandes.
  - `"liblinear"`: Bueno para datasets pequeños.
  - `"saga"`: Funciona bien con regularización `elasticnet`.

#### **Configuración de GridSearchCV**
```python
grid = GridSearchCV(model, hyperparams, scoring="accuracy", cv=10)
```
- **`GridSearchCV`**: Busca automáticamente la mejor combinación de hiperparámetros probando todas las combinaciones posibles.
- **`model`**: Es el modelo de regresión logística que estás optimizando.
- **`hyperparams`**: Son los hiperparámetros que definiste antes.
- **`scoring="accuracy"`**: Evalúa el modelo usando la métrica de precisión (qué tan bien clasifica los datos).
- **`cv=10`**: Realiza validación cruzada con 10 divisiones (entrena y prueba el modelo 10 veces con diferentes subconjuntos de datos).

---

### **GridSearchCV:**
1. Prueba todas las combinaciones de los hiperparámetros que definiste.
2. Evalúa el modelo con cada combinación usando validación cruzada.
3. Encuentra la combinación de hiperparámetros que da el mejor rendimiento (en este caso, la mayor precisión).


### **Resumen**
Este código:
- Usa **GridSearchCV** para optimizar el modelo de regresión logística.
- Encuentra los mejores valores para los hiperparámetros (`C`, `penalty`, `solver`).
- Evalúa el modelo con validación cruzada para garantizar que funcione bien con datos nuevos.
